```python
import cv2
import os
import numpy as np 
from PIL import Image 
from os.path import join
from datetime import datetime
```

```python
recognizer = cv2.face.LBPHFaceRecognizer_create() 
path = 'final' 
if not os.path.exists('../recognizer'):
    os.makedirs('../recognizer') 

def getImageswithId(path):
    faces = []
    faceid = []

    for root,directory,filenames in os.walk(path):
        for filename in filenames:
            id = os.path.basename(root) 
            img_path = os.path.join(root,filename)
            print('img_path:',img_path)
            print('id:',id)
            test_img = cv2.imread(img_path)
            if test_img is None:
                print('image not loaded poperly - cv2 cant read!!')
                continue
            gray_img=cv2.cvtColor(test_img,cv2.COLOR_BGR2GRAY)
            face_haar_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
            face=face_haar_cascade.detectMultiScale(gray_img,scaleFactor=1.32,minNeighbors=5)
            if len(face)!=1:
                continue 
            (x,y,w,h) = face[0]
            gray = gray_img[y:y+h,x:x+h]
            equ = cv2.equalizeHist(gray) 
            final = cv2.medianBlur(equ, 3)
            faces.append(final)
            faceid.append(int(id))
    return faceid,faces

faceid , faces = getImageswithId(path)
recognizer.train(faces,np.array(faceid))
recognizer.save('recognizer/trainingData.yml')
cv2.destroyAllWindows()
```

```python
faceDetect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
cam = cv2.VideoCapture(0)

rec=cv2.face.LBPHFaceRecognizer_create()
rec.read("recognizer\\trainingData.yml")

id=0
font = cv2.FONT_HERSHEY_SIMPLEX

while(cam.isOpened()):    
    ret, img = cam.read()
    if ret == True:
        gray=cv2.cvtColor(img,cv2.COLOR_BGRA2GRAY)
        equ = cv2.equalizeHist(gray) 
        final = cv2.medianBlur(equ, 3)
        faces = faceDetect.detectMultiScale(final,1.3,5)
        for (x,y,w,h) in faces:
            cv2.rectangle(img, (x,y), (x+w , y+h), (0,255,0), 2)
            id,conf=rec.predict(gray[y:y+h,x:x+w])
            now=datetime.now()
            time=now.strftime('%H:%M:%S')
            cv2.putText(img,str(id), (x,y+h),font,4,(255,255,0),4,cv2.LINE_AA)
            cv2.putText(img,time,(50, 50),font, 1,(0,0,0),2,cv2.LINE_4) 
        cv2.imshow('Face' , img)
        if(cv2.waitKey(1)==ord('q')):
            break
    else:
        break

cam.release()
cv2.destroyAllWindows()
```
